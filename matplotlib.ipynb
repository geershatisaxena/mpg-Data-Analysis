#### **Pie Charts**

#### **MPG Category Pie (Using Bins)**

```python
mpg_bins = pd.cut(df["mpg"], bins=3, labels=["Low","Medium","High"])
plt.pie(mpg_bins.value_counts(),
        labels=mpg_bins.value_counts().index,
        colors=["red","yellow","green"],
        autopct="%1.1f%%")
plt.title("MPG Category Distribution")
```

```python
mpg_bins2 = pd.cut(df["mpg"], bins=4)
plt.pie(mpg_bins2.value_counts(),
        labels=mpg_bins2.value_counts().index,
        colors=["purple","cyan","orange","brown"],
        autopct="%1.1f%%")
plt.title("MPG 4-Level Distribution")
```

#### **Weight Category Pie**

```python
weight_bins = pd.qcut(df["weight"], q=4, labels=["Q1","Q2","Q3","Q4"])
plt.pie(weight_bins.value_counts(),
        labels=weight_bins.value_counts().index,
        colors=["pink","gray","olive","blue"],
        autopct="%1.1f%%")
plt.title("Weight Quartiles")
```

#### **Donut Style Pie Charts**

```python
plt.pie(df["origin"].value_counts(),
        labels=df["origin"].value_counts().index,
        colors=["red","blue","green"],
        autopct="%1.1f%%")
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Donut Chart - Origin")
```

```python
plt.pie(df["cylinders"].value_counts(),
        labels=df["cylinders"].value_counts().index,
        colors=["orange","purple","brown","pink","gray"],
        autopct="%1.1f%%")
centre_circle = plt.Circle((0,0),0.65,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Donut Chart - Cylinders")
```

#### **Exploded Pie Charts**

```python
explode = (0.1,0,0)
plt.pie(df["origin"].value_counts(),
        labels=df["origin"].value_counts().index,
        explode=explode,
        colors=["gold","teal","navy"],
        autopct="%1.1f%%")
plt.title("Exploded Origin Pie")
```

```python
explode = [0.1]*len(df["cylinders"].value_counts())
plt.pie(df["cylinders"].value_counts(),
        labels=df["cylinders"].value_counts().index,
        explode=explode,
        colors=["coral","lime","indigo","maroon","cyan"],
        autopct="%1.1f%%")
plt.title("Exploded Cylinder Pie")
```
